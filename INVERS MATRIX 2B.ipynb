## 1. Pengertian Invers Matriks

Misalkan *A* dan *B* adalah dua matriks persegi dengan ordo yang sama.  
Matriks *A* dan *B* dikatakan *saling invers* jika memenuhi hubungan:

**AB = BA = I**

Dimana:

- *A⁻¹* = invers matriks A  
- *I* = matriks identitas  

Contoh matriks identitas:

I =

| 1 | 0 |
|---|---|
| 0 | 1 |

Jika suatu matriks *A* dikalikan dengan inversnya *A⁻¹*, maka hasilnya adalah **matriks identitas**.

---

## 2. Syarat Matriks Memiliki Invers

Suatu matriks memiliki invers jika:

1. Matriks berbentuk *persegi (n × n)*
2. *Determinan matriks tidak sama dengan nol*

Jika determinan = 0 maka matriks disebut *singular* dan **tidak memiliki invers**.

---

# 3. Rumus Umum Invers Matriks

Rumus invers matriks adalah:

**A⁻¹ = (1 / det(A)) × Adj(A)**

Dimana:

- *det(A)* = determinan matriks  
- *Adj(A)* = adjoin matriks  

---

# 4. Invers Matriks Ordo 2 × 2

Jika

A =

| a | b |
|---|---|
| c | d |

maka

**det(A) = ad − bc**

dan inversnya adalah:

A⁻¹ = (1 / (ad − bc))

| d | −b |
|---|---|
| −c | a |

---

## Contoh 1

Diketahui

A =

| 3 | 5 |
|---|---|
| 1 | 2 |

B =

| 2 | -5 |
|---|---|
| -1 | 3 |

Hitung:

AB

AB =

| 3 | 5 |
|---|---|
| 1 | 2 |

×

| 2 | -5 |
|---|---|
| -1 | 3 |

=

| (3×2 + 5×-1) | (3×-5 + 5×3) |
|---|---|
| (1×2 + 2×-1) | (1×-5 + 2×3) |

=

| 1 | 0 |
|---|---|
| 0 | 1 |

Karena

**AB = BA = I**

maka

A = B⁻¹  
B = A⁻¹

---

## Contoh 2

Jika

A =

| 3 | 5 |
|---|---|
| 1 | 2 |

Langkah 1: Hitung determinan

det(A) = (3×2) − (5×1)

det(A) = **6 − 5 = 1**

Langkah 2: Gunakan rumus invers

A⁻¹ = 1/det(A)

| 2 | -5 |
|---|---|
| -1 | 3 |

A⁻¹ =

| 2 | -5 |
|---|---|
| -1 | 3 |

# 5. Invers Matriks Ordo 3 × 3

Jika matriks

B =

| 2 | 1 | 0 |
|---|---|---|
| 3 | -1 | 4 |
| 2 | 2 | 0 |

maka invers dihitung dengan rumus:

**B⁻¹ = (1 / det(B)) × Adj(B)**

---

## Langkah 1: Hitung Determinan

det(B) menggunakan metode *Sarrus*
# Menentukan Determinan Matriks Ordo 3×3 dengan Metode Sarrus

Diketahui matriks:

B =

| 2 | 1 | 0 |
|---|---|---|
| 3 | -1 | 4 |
| 2 | 2 | 0 |

---

## Langkah 1: Salin dua kolom pertama di sebelah kanan

| 2 | 1 | 0 | 2 | 1 |
|---|---|---|---|---|
| 3 | -1 | 4 | 3 | -1 |
| 2 | 2 | 0 | 2 | 2 |

---

## Langkah 2: Hitung diagonal utama (kiri atas → kanan bawah)

1. 2 × (-1) × 0 = 0  
2. 1 × 4 × 2 = 8  
3. 0 × 3 × 2 = 0  

Jumlah diagonal utama:

0 + 8 + 0 = **8**

---

## Langkah 3: Hitung diagonal kedua (kanan atas → kiri bawah)

1. 0 × (-1) × 2 = 0  
2. 1 × 3 × 0 = 0  
3. 2 × 4 × 2 = 16  

Jumlah diagonal kedua:

0 + 0 + 16 = **16**

---

## Langkah 4: Hitung determinan

det(B) = (jumlah diagonal utama) − (jumlah diagonal kedua)

det(B) = 8 − 16

det(B) = **−8**

## Langkah 2: Cari Minor
# Langkah Umum Mencari Minor

1. Pilih elemen yang akan dicari minornya  
2. Hapus baris tempat elemen berada  
3. Hapus kolom tempat elemen berada  
4. Hitung determinan matriks yang tersisa

---

# Contoh Perhitungan Minor

## 1. Minor M₁₁

Elemen posisi (1,1) adalah **2**

Hapus baris 1 dan kolom 1

Sisa matriks:

| -1 | 4 |
|----|---|
| 2  | 0 |

Hitung determinan:

(-1 × 0) − (4 × 2)

= 0 − 8

**M₁₁ = -8**

---

## 2. Minor M₁₂

Elemen posisi (1,2) adalah **1**

Hapus baris 1 dan kolom 2

Sisa matriks:

| 3 | 4 |
|---|---|
| 2 | 0 |

Hitung determinan:

(3 × 0) − (4 × 2)

= 0 − 8

**M₁₂ = -8**

---

## 3. Minor M₁₃

Elemen posisi (1,3) adalah **0**

Hapus baris 1 dan kolom 3

Sisa matriks:

| 3 | -1 |
|---|---|
| 2 | 2 |

Hitung determinan:

(3 × 2) − (-1 × 2)

= 6 + 2

**M₁₃ = 8**

---

# Contoh Minor Lain

## M₂₃

Hapus baris 2 dan kolom 3

| 2 | 1 |
|---|---|
| 2 | 2 |

Determinan:

(2 × 2) − (1 × 2)

= 4 − 2

**M₂₃ = 2**

Tanda kofaktor:

## Langkah 4: Adjoin Matriks

Adj(B) adalah *transpose dari kofaktor*
Dari baris ke kolom

Adj(B) =

| -8 | 0 | 4 |
|---|---|---|
| 8 | 0 | -8 |
| 8 | -2 | -5 |

---

## Langkah 5: Hitung Invers

B⁻¹ = (1 / -8)

| -8 | 0 | 4 |
|---|---|---|
| 8 | 0 | -8 |
| 8 | -2 | -5 |

Hasil:

B⁻¹ =

| 1 | 0 | -1/2 |
|---|---|---|
| -1 | 0 | 1 |
| -1 | 1/4 | 5/8 |

# 6. Ringkasan

| Konsep | Penjelasan |
|------|------|
| Invers matriks | matriks kebalikan |
| Syarat | matriks persegi dan det(A) ≠ 0 |
| Rumus umum | A⁻¹ = (1/det(A)) Adj(A) |
| Ordo 2×2 | tukar diagonal dan ubah tanda elemen lain |
| Ordo 3×3 | hitung determinan → kofaktor → adjoin |

# 7. Kegunaan Invers Matriks

Invers matriks digunakan untuk:

1. Menyelesaikan sistem persamaan linear  

   Ax = b  

   solusinya:

   **x = A⁻¹ b**

2. Komputasi matematika  
3. Analisis data  
4. Pemodelan fisika dan geofisika
